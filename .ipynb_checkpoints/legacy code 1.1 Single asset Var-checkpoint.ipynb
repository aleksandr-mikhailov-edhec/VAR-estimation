```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model
```

```python
base_colors = ["#14284F", "#EB008B", "#5CD6AE", "#FFB81C", "#C2AFF0"]
plt.rcParams["axes.prop_cycle"] = plt.cycler(
    color=base_colors 
)
```

```python
# Download S&P 500 data from Yahoo Finance
sp500 = yf.download('^GSPC', start='2002-01-01', end='2024-05-27')

# Calculate daily log returns
sp500['Log Return'] = np.log(sp500['Adj Close'] / sp500['Adj Close'].shift(1))

# Drop the first row with NaN value
sp500 = sp500.dropna()

sp500.index = pd.to_datetime(sp500.index)
```

```python
# Define the rolling window size (e.g., 252 trading days for 1 year)
rolling_window = 126 # we take 6 months

# Calculate rolling historical VaR at the 99% confidence level
confidence_level = 0.99

# Number of days to simulate ahead & assess
num_days = 10#21  
```

```python
sp500
```

```python
sp500 = sp500.copy()
sp500['Rolling VaR'] = sp500['Log Return'].rolling(window=num_days).sum(\
).rolling(window=rolling_window).quantile(1 - confidence_level)
sp500 = sp500.dropna()
```

```python
sp500.loc[:,'Exceedances of Rolling VaR'] = (sp500['Log Return'].rolling(window=num_days).sum() < sp500['Rolling VaR']).astype(np.int64)*sp500['Rolling VaR']
```

Monte Carlo Simultation

```python
# Monte Carlo simulation parameters
num_simulations = 4000  # Number of simulations


# Function to perform Monte Carlo simulation
def monte_carlo_var(log_returns, num_simulations, num_days, confidence_level):
    # Get the mean and standard deviation of the log returns
    mean_return = log_returns.mean()
    std_return = log_returns.std()


    # Simulate future returns
    simulated_returns = np.random.normal(mean_return, std_return, (num_simulations, num_days))

    # Calculate cumulative returns for each simulation
    cumulative_returns = simulated_returns.sum(axis=1)
    # Calculate the simulated VaR at the desired confidence level
    simulated_var = np.percentile(cumulative_returns, (1 - confidence_level) * 100)
    
    return simulated_var
```

```python
# Calculate rolling Monte Carlo VaR
sp500['Rolling Monte Carlo VaR'] = sp500['Log Return'].rolling(window=rolling_window).apply(
    lambda x: monte_carlo_var(x, num_simulations, num_days, confidence_level), raw=False)

sp500.loc[:,'Exceedances of Monte Carlo VaR'] = (sp500['Log Return'].rolling(window=num_days).sum() < \
                                                 sp500['Rolling Monte Carlo VaR']).astype(np.int64)*sp500['Rolling Monte Carlo VaR']
```

```python
plt.figure(figsize=(14, 7))
plt.plot(sp500['Log Return'].rolling(num_days).sum(), label='S&P 500 cum log returns ' + str(num_days),zorder=0)
plt.plot(sp500['Rolling VaR'], label='Rolling 1-Year 99% VaR', linestyle='--',zorder=1, alpha = 0.8)
plt.plot(sp500['Rolling Monte Carlo VaR'], label='Rolling 1-Year Monte Carlo VaR', linestyle='--', zorder=0.5, alpha = 0.8)

plt.scatter(sp500.loc[sp500['Exceedances of Rolling VaR']!=0,'Exceedances of Rolling VaR'].index,
            sp500.loc[sp500['Exceedances of Rolling VaR']!=0,'Exceedances of Rolling VaR'],
            color = 'red', alpha = 0.7,zorder=2, s = 8)

plt.scatter(sp500.loc[sp500['Exceedances of Monte Carlo VaR']!=0,'Exceedances of Monte Carlo VaR'].index,
            sp500.loc[sp500['Exceedances of Monte Carlo VaR']!=0,'Exceedances of Monte Carlo VaR'],
            color = 'darkblue', alpha = 0.7,zorder=1.5, s = 8)


plt.xlim(sp500.index[rolling_window],sp500.index[-1])
plt.title('S&P 500 log returns and Rolling 1-Year 99% VaR')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
real_failure_prc_mc = np.sum(sp500['Exceedances of Monte Carlo VaR'].dropna()!=0)/len(sp500['Exceedances of Monte Carlo VaR'].dropna()!=0)
real_failure_prc_roll = np.sum(sp500['Exceedances of Rolling VaR'].dropna()!=0)/len(sp500['Exceedances of Rolling VaR'].dropna()!=0)
real_failure_prc_mc.round(3),real_failure_prc_roll.round(3)
```

```python
from scipy.stats import binomtest

var_breaches = list(sp500.dropna().loc[sp500['Exceedances of Monte Carlo VaR'].dropna()!=0,:].index)

# Calculate total number of observations
total_observations = len(sp500.dropna())
expected_breaches = total_observations * (1-confidence_level)

# Calculate actual number of breaches
actual_breaches = len(var_breaches )

# Perform a binomial test
p_value = binomtest(actual_breaches, total_observations, 0.05)

print(f"Total Observations: {total_observations}")
print(f"Expected Breaches: {expected_breaches}")
print(f"Actual Breaches: {actual_breaches}")
print(f"P-Value: {p_value}")
```

```python
def monte_carlo_var_garch(log_returns, num_simulations, num_days, confidence_level):
    # Fit a GARCH(1,1) model to the log returns
    log_returns = log_returns*100
    garch_model = arch_model(log_returns, vol='Garch', p=1, q=1)
    garch_fit = garch_model.fit(disp='off')
    
    # Get the mean return and GARCH-based forecasted volatility
    mean_return = log_returns.mean()
    forecast = garch_fit.forecast(horizon=num_days)
    std_return = np.sqrt(forecast.variance.values[-1, :])
    
    # Simulate future returns for the given number of days
    simulated_returns = np.random.normal(mean_return, std_return, (num_simulations, num_days))

    # Calculate cumulative returns for each simulation over the specified number of days
    cumulative_returns = simulated_returns.sum(axis=1)

    # Calculate the simulated VaR at the desired confidence level
    simulated_var = np.percentile(cumulative_returns, (1 - confidence_level) * 100)

    return simulated_var/100

```

```python
# Calculate rolling Monte Carlo VaR with GARCH
sp500['Rolling Monte Carlo GARCH VaR'] = sp500['Log Return'].rolling(window=rolling_window).apply(
    lambda x: monte_carlo_var_garch(x, num_simulations, num_days, confidence_level), raw=False)
```

```python
sp500.loc[:,'Exceedances of Monte Carlo GARCH VaR'] = (sp500['Log Return'].rolling(window=num_days).sum() < \
                                                 sp500['Rolling Monte Carlo GARCH VaR']).astype(np.int64)*sp500['Rolling Monte Carlo GARCH VaR']
```
