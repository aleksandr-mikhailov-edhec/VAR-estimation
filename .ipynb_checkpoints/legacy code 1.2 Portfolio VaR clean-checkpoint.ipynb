# 2. Functions

```python
def data_initialisation(name_portfolio_weights_file,
                        name_portfolio_returns_file,
                        rolling_window):
    # weights have date de valeur
    weights = pd.read_excel(name_portfolio_weights_file,sheet_name='Allocations',
                            index_col = 1, header = 1).drop(columns = ['Date de saisie'])
    
    # while reading returns specify header and index columns; drop the latest values of the latest day 
    #(they are NaNs & fill all other latest NaNs through forward filling)
    # then fill by 100 any assets that have been implemented from not the first date
    returns = pd.read_excel(name_portfolio_returns_file,index_col = 0, header = 3).iloc[:-1,:].ffill().fillna(100)
    
    log_returns = np.log(returns).diff().fillna(0)
    returns = returns.pct_change().fillna(0)

    # initialise rebalancing dates in a specific way
    rebalancing_dates = pd.read_excel(name_portfolio_weights_file,sheet_name='Allocations',
                                  header = 1).loc[:,['Date de saisie','Date de valeur']]

    # delete non-existing dates
    rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de saisie']<=returns.index[-1],:]
    rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de valeur']<=returns.index[-1],:]
    
    rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de saisie']>=returns.index[rolling_window],:]
    rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de valeur']>=returns.index[rolling_window],:]
    
    rebalancing_dates = rebalancing_dates.reset_index(drop = True)

    return weights, returns, log_returns, rebalancing_dates
```

```python
def calculate_weight_drift(monthly_weights, daily_returns):
    # function for calculation of weight drift between rebalancing dates.
    # Give it your weights on Date de Valeur & daily returns of assets. 
    # It will create 
    
    # Initialize the weights DataFrame with daily index
    daily_weights = pd.DataFrame(index = daily_returns.index, columns = daily_returns.columns)

    if monthly_weights.index[0] != daily_returns.index[0]:
        raise ValueError('first date for monthly_weights and daily_returns should be same')
        
    # Start with the initial monthly weights
    current_weights = monthly_weights.iloc[0]
    daily_weights.loc[daily_returns.index[0]] = current_weights
    
    for i, date in enumerate(daily_returns.index[1:]):
        # Calculate the portfolio value based on the current weights and previous day returns
        current_weights = current_weights * (1 + daily_returns.iloc[i-1])
        
        # Normalize the weights so that they sum to 1
        current_weights = current_weights / current_weights.sum()
        
        # If a new month starts, update the current weights to the next monthly weights
        if date in monthly_weights.index:
            current_weights = monthly_weights.loc[date]

        # Store the calculated weights for the current date
        daily_weights.loc[date] = current_weights
        
    return daily_weights
```

```python
def monte_carlo_var_garch(log_returns, num_simulations, num_days, confidence_level,scaler = 1000):
    '''
    This function calculates GARCH-fitted monte-carlo simulation of log-returns (normal distr)
    and finds its VaR
    '''
    # scaler is needed to avoid one warning related to size of y
    
    # Fit a GARCH(1,1) model to the log returns
    log_returns = log_returns * scaler
    garch_model = arch_model(log_returns, vol='Garch', p=1, q=1)
    garch_fit = garch_model.fit(disp='off')
    
    # Get the mean return and GARCH-based forecasted volatility
    mean_return = log_returns.mean()
    forecast = garch_fit.forecast(horizon=num_days)
    std_return = np.sqrt(forecast.variance.values[-1, :])
    
    # Simulate future returns for the given number of days
    simulated_returns = np.random.normal(mean_return, std_return, (num_simulations, num_days))

    # Calculate cumulative returns for each simulation over the specified number of days
    cumulative_returns = simulated_returns.sum(axis=1)

    # Calculate the simulated VaR at the desired confidence level
    simulated_var = np.percentile(cumulative_returns, (1 - confidence_level) * 100)

    return simulated_var / scaler
```

```python
# STRATEGY 2

#!!! Uses previous functions
def desinvest_in_risky_asset_strategy(vars_assets,
                                      rebalancing_dates,
                                      weights,
                                      returns,
                                      strategy_results,
                                      strategy_name = 'sample risky asset VaR desinvestment strategy',
                                      var_treshold = 0.1,
                                      cash_asset_name = 'AXA CT A Capitalisation EUR'):

    # For each rebalancing date we check each asset in portfolio. 
    # If this asset is breaching certain VaR, weights for the next month are spread among other assets (that did not breah their VaRs)

    
    # find True/False matrixes of breaching VaRs
    # if VaR is NOT breached, we are agreed to invest in an asset => pose True
    var_is_at_fine_level = vars_assets >= np.log(1-var_treshold)

    # i used the 1st year to fit models, so some first dates do not  exist
    weights_for_backtest = weights.loc[rebalancing_dates['Date de valeur'],:]
    
    # I add a very tiny weight to cash if we did not invest in it to avoid situations when we 'invest in nothing'
    weights_for_backtest.loc[weights_for_backtest[cash_asset_name] == 0,cash_asset_name] = 0.0001

    # rename index. FromDate de sassie to date de valeur
    # on the next step we will need it for implementation of weights => need actual dates
    var_is_at_fine_level.index = rebalancing_dates['Date de valeur']
    
    # decision whether to desinvest in asset or not. It is taken on 
    # Date de sessie by var_is_at_fine_level before reindexing
    # as plus, if actual weight on Date de valeur was not 0, the asset
    # is really used.
    # I index it as 'Date de valeur', because at the next step will need actual implementation
    # date to change weights
    decision_matrix = (weights_for_backtest!=0)*(var_is_at_fine_level)

    # all assets' weights that had weight of 0 at 'Date de valeur' or were bad by var
    # at 'Date de sassie' are set to 0
    weight_matrix = weights_for_backtest*decision_matrix

    # sum of weights is normalized to 1
    weight_matrix = weight_matrix.div(weight_matrix.sum(axis = 1),axis = 0)

    # make daily weights 
    daily_weights = calculate_weight_drift(weight_matrix, returns.loc[weight_matrix.index[0]:,:])

    
    strategy_returns = (daily_weights * returns.loc[daily_weights.index,:]).sum(axis = 1)

    strategy_results[strategy_name + str(var_treshold)] = strategy_returns
    return strategy_results
```

```python
# STRATEGY 1

def test_strategy_invest_desinvest(log_returns_of_asset,
                                   var_of_asset, 
                                   rebalancing_dates,
                                   strategy_results,                # df for returns of strategies
                                   treshold_of_desinvestment = 0.1, # in usual %
                                   strategy_name = 'sample invest/desinvet strategy'
                                  ):
    # Every rebalancing date (Date de saisie) that was specified by the company we test GARCH fitted Monte Carlo VaR.
    # If this Var shows that breaches a certain treshold for the next 30 days,
    # we do not invest from the closest Date de valeur to the second closest.
    # We will test several tresholds and test them VS our real portfolio.
    # All portfolios will be tested for 95% VaR that was fitted on the last year of value 251
    
    desinvestment_indicator = var_of_asset < np.log(1 - treshold_of_desinvestment)

    for i in range(len(rebalancing_dates['Date de saisie'][:-1])):
    
        current_date = rebalancing_dates['Date de saisie'][i]
        
        date_of_decision_implementation = rebalancing_dates['Date de valeur'][i]
        date_till_which_decision_is_implemented = rebalancing_dates['Date de valeur'][i+1]

        log_returns_for_the_period = log_returns_of_asset[date_of_decision_implementation:date_till_which_decision_is_implemented]
        
        if desinvestment_indicator[current_date]:
            # desinvest
            strategy_results.loc[date_of_decision_implementation:date_till_which_decision_is_implemented, strategy_name] = 0
        else:
            # invest
            # return actual returns; not log-returns
            strategy_results.loc[date_of_decision_implementation:date_till_which_decision_is_implemented,
            strategy_name] = np.exp(log_returns_for_the_period)-1


    # run the final cycle
    current_date = rebalancing_dates['Date de saisie'][i+1]
    date_of_decision_implementation = rebalancing_dates['Date de valeur'][i+1]
    date_till_which_decision_is_implemented = log_returns_of_asset.index[-1]

    log_returns_for_the_period = log_returns_of_asset[date_of_decision_implementation:date_till_which_decision_is_implemented]
    
    if desinvestment_indicator[current_date]:
        strategy_results.loc[date_of_decision_implementation:date_till_which_decision_is_implemented, strategy_name] = 0
    else:

        strategy_results.loc[date_of_decision_implementation:date_till_which_decision_is_implemented,
        strategy_name] = np.exp(log_returns_for_the_period)-1    
    
    return strategy_results
```

```python
# !!! Calls another function
def find_var_of_all_assets(log_returns,
                           rebalancing_dates,
                           random_number,
                           cash_asset_name,
                           num_simulations, 
                           num_days, 
                           confidence_level,
                           scaler
                          ):
    
    vars_assets = pd.DataFrame(columns = log_returns.columns, index = rebalancing_dates['Date de saisie'])
    
    for asset in tqdm(log_returns.columns):
    
        # exclude cash asset from var caculation. We assume its var is always 0
        if asset != cash_asset_name:
            
            for date in rebalancing_dates['Date de saisie']:
                
                np.random.RandomState(random_number)
                if log_returns.loc[:date,asset].iloc[-rolling_window:].sum() == 0:
                    
                    # warnings appear because of 'EDR EDR Short Duration Credit A USD'
                    # => if return 0, VaR is not applied
                   vars_assets.loc[date,asset] = 0
                    
                else:
                    vars_assets.loc[date,asset] = monte_carlo_var_garch(log_returns.loc[:date,asset].iloc[-rolling_window:],
                                                                        num_simulations, num_days, confidence_level,scaler)
    
    # set cash asset to VaR 0
    vars_assets.loc[:,cash_asset_name] = 0
    
    return vars_assets
```

```python
def assess_strategies(strategy_results, risk_free_rate = 0.02):
    results = pd.DataFrame(columns = ['annualized return', 'annualized std','max drawdown','SR '+str(risk_free_rate)])

    for portfolio in strategy_results.columns:
        std_ = strategy_results[portfolio].std()*np.sqrt(251)
        ret_ = (1+strategy_results[portfolio]).prod()**(strategy_results.shape[0]/251)
        portfolio_growth = (1+strategy_results[portfolio]).cumprod()
        
        results.loc[portfolio,'annualized std'] = round(std_,3)
        results.loc[portfolio,'annualized return'] = round(ret_,3)
        results.loc[portfolio,'max drawdown'] = round(np.max(portfolio_growth.expanding().max() - portfolio_growth.expanding().min()),3)
        results.loc[portfolio,'SR '+str(risk_free_rate)] = round((ret_-risk_free_rate) / std_,3)
        
    return results
```
