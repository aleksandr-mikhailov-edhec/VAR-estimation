```python
# Define a function to fit a t-distribution to the data
def fit_t_distribution(data):
    # Function to minimize
    def negative_log_likelihood(params):
        df, loc, scale = params[0], params[1], params[2]
        return -np.sum(t.logpdf(data, df, loc, scale))

    # Initial guesses for df, loc, and scale
    initial_params = [2, np.mean(data), np.std(data)]
    bounds = [(1, 30), (None, None), (0.001, None)]
    
    result = minimize(negative_log_likelihood, initial_params, bounds=bounds)
    df, loc, scale = result.x
    return df, loc, scale

# Plot QQ plots and test goodness-of-fit
plt.figure(figsize=(20, 30))
goodness_of_fit_results = {}

for i, column in enumerate(returns.columns):
    
    ret_of_asset = log_returns[column]
    ret_of_asset = ret_of_asset[ret_of_asset != 0]
    
    df_t, loc, scale = fit_t_distribution(ret_of_asset)
    
    # Create QQ plot
    plt.subplot(10, 4, i+1)
    stats.probplot(ret_of_asset, dist="t", sparams=(df_t,), plot=plt)
    #(osm, osr), (slope, intercept, r) = stats.probplot(ret_of_asset, dist="norm", plot=None)
    #plt.plot(osm, osr, 'go', label="Normal Data Points")
    #plt.plot(osm, slope*osm + intercept, 'g-', label="Normal Fit", alpha = 0.7)
    plt.title(f'{column} QQ Plot')

    # Goodness-of-fit test
    d_stat, p_value = kstest(ret_of_asset, 't', args=(df_t, loc, scale))
    goodness_of_fit_results[column] = {'DF': df_t, 'Location': loc, 'Scale': scale, 'KS Statistic': d_stat, 'p-value': p_value}

plt.tight_layout()
plt.show()
```

**Result: they are definitely not normally distributed, but t distribution looks fine**

# 2. Make portfolio

```python
# function for calculation of weight drift between rebalancing dates
def calculate_weight_drift(monthly_weights, daily_returns):
    # Initialize the weights DataFrame with daily index
    daily_weights = pd.DataFrame(index=daily_returns.index, columns=daily_returns.columns)

    if monthly_weights.index[0] != daily_returns.index[0]:
        raise ValueError('first date for monthly_weights and daily_returns should be same')
    # Start with the initial monthly weights
    current_weights = monthly_weights.iloc[0]
    daily_weights.loc[daily_returns.index[0]] = current_weights
    
    for i, date in enumerate(daily_returns.index[1:]):
        # Calculate the portfolio value based on the current weights and previous day returns
        current_weights = current_weights * (1 + daily_returns.iloc[i-1])
        
        # Normalize the weights so that they sum to 1
        current_weights = current_weights / current_weights.sum()
        
        # If a new month starts, update the current weights to the next monthly weights
        if date in monthly_weights.index:
            current_weights = monthly_weights.loc[date]

        # Store the calculated weights for the current date
        daily_weights.loc[date] = current_weights
        
    return daily_weights
```

```python
# Define the rolling window size (e.g., 252 trading days for 1 year)
rolling_window = 251 # we take 6 months

# Calculate rolling historical VaR at the 99% confidence level
confidence_level = 0.95

# Number of days to simulate ahead & assess
# for us it is 21 + 9 (1 month + lag of strategy implementation)
num_days = 30 

var = pd.DataFrame(portfolio_returns, columns = ['portfolio returns'])
var['log portfolio returns'] = np.log((var['portfolio returns']+1).astype(np.float64))
var['Rolling Historical VaR'] = var['log portfolio returns'].rolling(window=num_days).sum(\
).rolling(window=rolling_window).quantile(1 - confidence_level)
var
```

```python
# Monte Carlo simulation parameters
num_simulations = 5000  # Number of simulations


# Function to perform Monte Carlo simulation
def monte_carlo_var(log_returns, num_simulations, num_days, confidence_level):
    # Get the mean and standard deviation of the log returns
    mean_return = log_returns.mean()
    std_return = log_returns.std()


    # Simulate future returns
    simulated_returns = np.random.normal(mean_return, std_return, (num_simulations, num_days))

    # Calculate cumulative returns for each simulation
    cumulative_returns = simulated_returns.sum(axis=1)
    # Calculate the simulated VaR at the desired confidence level
    simulated_var = np.percentile(cumulative_returns, (1 - confidence_level) * 100)
    
    return simulated_var
```

```python
def monte_carlo_var_garch(log_returns, num_simulations, num_days, confidence_level,scaler = 1000):
    # scaler is needed to avoid one warning related to size of y
    
    # Fit a GARCH(1,1) model to the log returns
    log_returns = log_returns * scaler
    garch_model = arch_model(log_returns, vol='Garch', p=1, q=1)
    garch_fit = garch_model.fit(disp='off')
    
    # Get the mean return and GARCH-based forecasted volatility
    mean_return = log_returns.mean()
    forecast = garch_fit.forecast(horizon=num_days)
    std_return = np.sqrt(forecast.variance.values[-1, :])
    
    # Simulate future returns for the given number of days
    simulated_returns = np.random.normal(mean_return, std_return, (num_simulations, num_days))

    # Calculate cumulative returns for each simulation over the specified number of days
    cumulative_returns = simulated_returns.sum(axis=1)

    # Calculate the simulated VaR at the desired confidence level
    simulated_var = np.percentile(cumulative_returns, (1 - confidence_level) * 100)

    return simulated_var / scaler
```

```python
np.random.RandomState(42)
var['Rolling Monte Carlo VaR'] = var['log portfolio returns'].rolling(window=rolling_window).apply(
    lambda x: monte_carlo_var(x, num_simulations, num_days, confidence_level), raw=False)
```

```python
np.random.RandomState(42)
var['Rolling Monte Carlo GARCH VaR'] = var['log portfolio returns'].rolling(window=rolling_window).apply(
    lambda x: monte_carlo_var_garch(x, num_simulations, num_days, confidence_level), raw=False)
```

# 4. Assess simple strategy


- Every rebalancing date (Date de saisie) that was specified by the company we test GARCH fitted Monte Carlo VaR.
- If this Var shows that breaches a certain treshold for the next 30 days, we do not invest from the closest Date de valeur to the second closest.
- We will test several tresholds and test them VS our real portfolio.
- All portfolios will be tested for 95% VaR that was fitted on the last year of value 251

```python
rebalancing_dates = pd.read_excel(name_portfolio_weights_file,sheet_name='Allocations',
                                  header = 1).loc[:,['Date de saisie','Date de valeur']]
```

```python
var = var.dropna()
var = var.copy()

# delete non-existing dates
rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de saisie']<=var.index[-1],:]
rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de valeur']<=var.index[-1],:]

rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de saisie']>=var.index[0],:]
rebalancing_dates = rebalancing_dates.loc[rebalancing_dates['Date de valeur']>=var.index[0],:]

rebalancing_dates = rebalancing_dates.reset_index(drop = True)
```

```python
var.loc[:,'Rolling Monte Carlo GARCH VaR 5%'] = var['Rolling Monte Carlo GARCH VaR'] < np.log(1-0.05)
var.loc[:,'Rolling Monte Carlo GARCH VaR 10%'] = var['Rolling Monte Carlo GARCH VaR'] < np.log(1-0.1)
var.loc[:,'Rolling Monte Carlo GARCH VaR 15%'] = var['Rolling Monte Carlo GARCH VaR'] < np.log(1-0.15)
```

```python
def test_strategy(var, strategy_name):
    for i in range(len(rebalancing_dates['Date de saisie'][:-1])):
    
        current_date = rebalancing_dates['Date de saisie'][i]
        
        date_of_decision_implementation = rebalancing_dates['Date de valeur'][i]
        date_till_which_decision_is_implemented = rebalancing_dates['Date de valeur'][i+1]
        
        log_returns_for_the_period = list(var.loc[date_of_decision_implementation:date_till_which_decision_is_implemented,
        'log portfolio returns'])
    
        if var.loc[current_date,strategy_name]:
            # desinvest
            var.loc[date_of_decision_implementation:date_till_which_decision_is_implemented,
            strategy_name + ' strategy log returns'] = 0
        else:
            #invest
            var.loc[date_of_decision_implementation:date_till_which_decision_is_implemented,
            strategy_name + ' strategy log returns'] = log_returns_for_the_period
    return var
```

```python
var = test_strategy(var, 'Rolling Monte Carlo GARCH VaR 5%')
var = test_strategy(var, 'Rolling Monte Carlo GARCH VaR 10%')
var = test_strategy(var, 'Rolling Monte Carlo GARCH VaR 15%')
```

# 5. Assess more advanced strategy

- For each rebalancing date we check each asset in portfolio. If this asset is breaching certain VaR, weights for the next month are spread among other assets (that did not breah their VaRs)

It is too long to run model for each data point => we will run it only for rebalancing dates

```python
from tqdm import tqdm
```

```python
vars_assets = pd.DataFrame(columns = log_returns.columns, index = rebalancing_dates['Date de saisie'])
for asset in tqdm(log_returns.columns):

    # exclude cash asset from var caculation. We assume its var is always 0
    if asset != 'AXA CT A Capitalisation EUR':
        
        for date in rebalancing_dates['Date de saisie']:
            np.random.RandomState(42)
            if log_returns.loc[:date,asset].iloc[-rolling_window:].sum() == 0:
                # warnings appear because of 'EDR EDR Short Duration Credit A USD'
                # this asset's data does not exist till some moment
               vars_assets.loc[date,asset] = 0 
            else:
                vars_assets.loc[date,asset] = monte_carlo_var_garch(log_returns.loc[:date,asset].iloc[-rolling_window:],
                                                                    num_simulations, num_days, confidence_level)

# set cash asset to VaR 0
vars_assets.loc[:,'AXA CT A Capitalisation EUR'] = 0
```

```python
# rename index. FromDate de sassie to date de valeur
# on the next step we will need it for implementation of weights => need actual dates
vars_assets.index = rebalancing_dates['Date de valeur']

# find True/False matrixes of breaching VaRs
# if VaR is NOT breached, we are agreed to invest in an asset => pose True
vars_assets_5 = vars_assets >= np.log(1-0.05)
vars_assets_10 = vars_assets >= np.log(1-0.1)
vars_assets_15 = vars_assets >= np.log(1-0.15)
vars_assets_20 = vars_assets >= np.log(1-0.20)
```

Now need to update the table with initial weights & recalculate weights for each day based on their drifts.

If weight of an asset was 0, we cannot change it. If weight of an asset breaches var, we exclude it and weight all other assets more equally. If all assets breached VaR, we invest all in cash ['AXA CT A Capitalisation EUR']. For that reason I remade its VaR equal to 0 always (that would be otherwise still 0)

```python
# i used the 1st year to fit models, so some first dates do not  exist
weights_for_backtest = weights.loc[vars_assets.index,:]

# I add a very tiny weight to cash if we did not invest in it to avoid situations when we 'invest in nothing'
weights_for_backtest.loc[weights_for_backtest['AXA CT A Capitalisation EUR'] == 0,'AXA CT A Capitalisation EUR'] = 0.0001

# We invest in asset if VaR is ok & if we did invest in that asset in the past in realisty. I cannot invest in smth
# that was not in portfolio. I asslow the strategy to only close positions
decision_matrix_5 = (weights_for_backtest!=0)*(vars_assets_5)
decision_matrix_10 = (weights_for_backtest!=0)*(vars_assets_10)
decision_matrix_15 = (weights_for_backtest!=0)*(vars_assets_15)
decision_matrix_20 = (weights_for_backtest!=0)*(vars_assets_15)

display(weights_for_backtest.head(2))
display(vars_assets_10.head(2))
display(decision_matrix_10.head(2))
```

```python
# create matrics with new weights
weight_matrix_5 = weights_for_backtest*decision_matrix_5
weight_matrix_10 = weights_for_backtest*decision_matrix_10
weight_matrix_15 = weights_for_backtest*decision_matrix_15
weight_matrix_20 = weights_for_backtest*decision_matrix_20

# show some summed weights:
display(weight_matrix_10.sum(axis = 1).head(10))
```

```python
# normalise weights according to the new rules
weight_matrix_5 = weight_matrix_5.div(weight_matrix_5.sum(axis = 1),axis = 0)
weight_matrix_10 = weight_matrix_10.div(weight_matrix_10.sum(axis = 1),axis = 0)
weight_matrix_15 = weight_matrix_15.div(weight_matrix_15.sum(axis = 1),axis = 0)
weight_matrix_20 = weight_matrix_20.div(weight_matrix_20.sum(axis = 1),axis = 0)

# show remade weights:
display(weight_matrix_10.sum(axis = 1).head(10))
```

```python
# make a portfolio weights daily based on the previous code => get portfolio weighted returns
daily_weights_5 = calculate_weight_drift(weight_matrix_5, returns.loc[weight_matrix_5.index[0]:,:])
daily_weights_10 = calculate_weight_drift(weight_matrix_10, returns.loc[weight_matrix_10.index[0]:,:])
daily_weights_15 = calculate_weight_drift(weight_matrix_15, returns.loc[weight_matrix_15.index[0]:,:])
daily_weights_20 = calculate_weight_drift(weight_matrix_20, returns.loc[weight_matrix_20.index[0]:,:])
```

```python
all_portfolios = pd.DataFrame(index = daily_weights_5.index)
all_portfolios['base portfolio'] = portfolio_returns[daily_weights_5.index[0]:]
all_portfolios['asset VaR 5 portfolio'] = (daily_weights_5 * returns[daily_weights_5.index[0]:]).sum(axis = 1)
all_portfolios['asset VaR 10 portfolio'] = (daily_weights_10 * returns[daily_weights_10.index[0]:]).sum(axis = 1)
all_portfolios['asset VaR 15 portfolio'] = (daily_weights_15 * returns[daily_weights_15.index[0]:]).sum(axis = 1)
all_portfolios['asset VaR 20 portfolio'] = (daily_weights_20 * returns[daily_weights_20.index[0]:]).sum(axis = 1)
all_portfolios['VaR 5 on portfolio'] = np.exp(var['Rolling Monte Carlo GARCH VaR 5% strategy log returns'])[daily_weights_5.index[0]:]-1
all_portfolios['VaR 10 on portfolio'] = np.exp(var['Rolling Monte Carlo GARCH VaR 10% strategy log returns'])[daily_weights_5.index[0]:]-1
#all_portfolios['VaR 15 on portfolio'] = np.exp(var['Rolling Monte Carlo GARCH VaR 15% strategy log returns'])[daily_weights_5.index[0]:]-1
```
